# T-Test of Data

```python
from scipy import stats
```

```python
churn = customers[customers['class']==1] 
```

```python
nonchurn = customers[customers['class']==0] 
```

```python
print(churn)
print(nonchurn)
```

# Significant Variables

```python
for item in numeric_variable:
    a = stats.ttest_ind(churn[item], nonchurn[item],  equal_var = False)
    if (a.pvalue<0.05):
        print(item)
```
